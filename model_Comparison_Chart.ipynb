# Comparison both of model KNN vs RoBERTS transformer

```python
# Import necessary libraries for plotting
import matplotlib.pyplot as plt
import numpy as np

# Define the model names and their scores (use your actual numbers)
models = ["KNN (Traditional ML)", "RoBERTa (Transformer)"]
accuracy_scores = [0.70, 0.94]
f1_scores = [0.41, 0.94]  # Macro average F1-score for each

# Bar chart setup
bar_width = 0.35
x = np.arange(len(models))

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))

# Plot bars for accuracy and F1-score
acc_bars = ax.bar(
    x - bar_width / 2, accuracy_scores, bar_width, label="Accuracy", color="#4c72b0"
)
f1_bars = ax.bar(
    x + bar_width / 2, f1_scores, bar_width, label="F1-score", color="#55a868"
)


# Function to add value labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(
            f"{height:.2f}",
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 3),  # 3 points vertical offset
            textcoords="offset points",
            ha="center",
            va="bottom",
        )


add_labels(acc_bars)
add_labels(f1_bars)

# Add labels, title, and grid
ax.set_ylabel("Score")
ax.set_title("Model Performance Comparison: KNN vs RoBERTa")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.ylim(0, 1.1)
plt.tight_layout()
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()
```
